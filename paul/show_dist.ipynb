```python
import matplotlib.pyplot as plt
import json
def show(filename,text):
    with open(filename,'r') as f:
        R=json.load(f)
    plt.plot(R['lang'],
    R['ndcg'],label=f'{text} nDCG@100')
    plt.legend()
    plt.grid()
    plt.ylim(0,1)
    print(f'{filename} Recall@100\t'+str(round(R['Avg']['recall'],4)))
    print(f'{filename} nDCG@10   \t'+str(round(R['Avg']['ndcg'],4)))

plt.figure()
show('../mdpr_bm25_0.13_dev.json')
show('../mdpr_tydi_qld_by_model_dev.json')
show('../temp_dev.json')
plt.show()
```
