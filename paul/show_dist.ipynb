```python
import matplotlib.pyplot as plt
import json
with open('../can_neg_dev.json','r') as f:
    R=json.load(f)
plt.figure()
plt.title('SVM hybrid all.')
plt.plot(R['lang'],
R['recall'],label='Recall@100')
plt.plot(R['lang'],
R['ndcg'],label='nDCG@100')
plt.legend()
plt.grid()
plt.ylim(0,1)
print('Recall@100\t'+str(round(R['Avg']['recall'],4)))
print('nDCG@10   \t'+str(round(R['Avg']['ndcg'],4)))
plt.show()
```
